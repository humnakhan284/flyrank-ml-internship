## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

## My lane as an ML task (Type)

This is a **classification** problem.

The goal is to classify each content page as either **declining** or **not declining** based on its historical performance metrics.

Classification is suitable because the output has two classes (0 = not declining, 1 = declining). The prediction helps identify pages that may need content updates.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or Proxy

The target is **is_declining_label**.

It is created from the observed value **trend_direction**.

If `trend_direction` is **down**, then:

- 1 = Declining
- 0 = Not Declining

This is a rule-based label created from the observed trend in the data.

```python
import pandas as pd

df = pd.read_csv(
    "/content/flyrank-ml-internship/data/raw/content_refresh_anonymized.csv"
)

df.head()
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success Metric

I will use **Precision@50** as the main evaluation metric.

Precision@50 measures how many of the top 50 pages predicted as declining are actually declining.

A higher Precision@50 means the model recommends better pages for review and content updates.

```python
# Create target label
df["is_declining_label"] = (
    df["impressions_last_30d"] < 0.8 * df["impressions_prev_30d"]
).astype(int)

# Check the label
print("Declining pages:", df["is_declining_label"].sum())
print("Total pages:", len(df))
print("Declining rate:", round(df["is_declining_label"].mean(), 3))
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML Beats a Fixed Rule

A simple rule such as "update every page older than 180 days" is not enough because many factors affect content performance.

Machine learning can learn patterns from multiple features like impressions, CTR, average position, and content age at the same time.

This helps make better decisions than using one fixed rule.

```python
features = [
    "content_age_days",
    "days_since_last_update",
    "impressions_90d",
    "avg_position",
    "ctr"
]

df[features].head()
```
